```python
def to_fc(image, name):
    return image.reduceRegions(
        collection=ee.FeatureCollection([ee.Feature(kolkata)]),
        reducer=ee.Reducer.mean(),
        scale=500
    ).map(lambda f: f.set('dataset', name))
```

```python
treecover2003_fc = to_fc(treecover2003, 'tree2003')
treecover2020_fc = to_fc(treecover2020, 'tree2020')

LAI2003_fc = to_fc(LAI2003, 'lai2003')
LAI2025_fc = to_fc(LAI2025, 'lai2025')

LST2003_fc = to_fc(LST2003, 'lst2003')
LST2025_fc = to_fc(LST2025, 'lst2025')
```

```python
aoi = kolkata

geemap.ee_export_image(treecover2003, filename='treecover2003.tif', scale=30, region=aoi)
geemap.ee_export_image(treecover2020, filename='treecover2020.tif', scale=30, region=aoi)

geemap.ee_export_image(LAI2003, filename='LAI2003.tif', scale=500, region=aoi)
geemap.ee_export_image(LAI2025, filename='LAI2025.tif', scale=500, region=aoi)

geemap.ee_export_image(LST2003, filename='LST2003.tif', scale=1000, region=aoi)
geemap.ee_export_image(LST2025, filename='LST2025.tif', scale=1000, region=aoi)
```
